<h2>Zadatak 6</h2>

Pronaći odstupanje svakog mjeseca od prosjeka tog mjeseca kroz godine

Npr. koliko siječanj 1951. odstupa od prosjeka (svih siječanja)?

```python
df.head()
```

```python
df["odst"]=0   #na ovaj način se kreira novu stupac i popunjava se u svakoj ćeliji sa vrijednošću 0
```

```python
df.head()
```

```python
for i in range (1,13):
    df.odst = np.max( abs(np.average(df[df.mon==i].avg) - df[i].avg) )
```

```python
df
```

```python
prosjeci=df.groupby("mon").aggregate(np.mean).avg.values

prosjeci[0]

def f_odst(m,a):
        return (a-prosjeci[m-1])
```

```python
df.odst=f_odst(m=df.mon,a=df.avg)
```
